```python
#imported items
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#upload data set
data = pd.read_csv('all_accelerometer_data_pids_13.csv')
print(data.head())
```

```python
# Plot for col x
plt.figure(figsize=(5, 5))
sns.histplot(data['x'], kde=True, bins=20)
plt.title('Histogram of x')
plt.show()
```

```python
# Plot for col y
plt.figure(figsize=(5, 5))
sns.histplot(data['y'], kde=True, bins=20)
plt.title('Histogram of y')
plt.show()
```

```python
# Plot for col z
plt.figure(figsize=(5, 5))
sns.histplot(data['z'], kde=True, bins=20)
plt.title('Histogram of z')
plt.show()
```
